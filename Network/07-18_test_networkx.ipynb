```python
pos = {
    'PHY183': np.array([0, 0.15]),
    'PHY184': np.array([0.25, 0.15]),
    'PHY193H': np.array([0, 0.85]),
    'PHY294H': np.array([0.25, 0.85]),
    'PHY215': np.array([0.4, 0.5]),
    'PHY321': np.array([0.65, 0.5]),
    'PHY410': np.array([0.8, 0.9]),
    'PHY471': np.array([1.0, 0.5]),
    'PHY481': np.array([0.8, 0.1])}
```

```python
Edge_labels = {('PHY183', 'PHY184'): 1500,
    ('PHY183', 'PHY294H'): 200,
    ('PHY193H', 'PHY184'): 300,
    ('PHY193H', 'PHY294H'): 400,
    ('PHY184', 'PHY215'): 1000,
    ('PHY294H', 'PHY215'): 300,
    ('PHY215', 'PHY321'): 600,
    ('PHY321', 'PHY471'): 300,
    ('PHY321', 'PHY410'): 200,
    ('PHY321', 'PHY481'): 100,
    ('PHY410', 'PHY471'): 100,
    ('PHY471', 'PHY410'): 200
}
```

```python
Edge_labels[('PHY183', 'PHY184')]
```

```python
Edge_width = np.array([15, 2, 3, 4, 10, 3, 6, 3, 2, 1, 1, 2])
```

```python
plt.figure(figsize=(18, 8))

graph = nx.MultiDiGraph()

for c in connections:
    graph.add_edge(c[0], c[1])
    
#pos = nx.spring_layout(graph)

nx.draw_networkx_nodes(graph, pos, node_size=3500, node_color='cyan')
nx.draw_networkx_edges(graph, pos, width=Edge_width, edge_color='black', arrows=True)
nx.draw_networkx_labels(graph, pos)
nx.draw_networkx_edge_labels(graph, pos, edge_labels=Edge_labels, label_pos=0.35)

plt.axis('off')
plt.show()
```

```python
a=np.reshape(np.random.randint(0, 2, size=25),(5,5))
D=nx.DiGraph(a) 
pos = nx.spectral_layout(D)
nx.draw_networkx_nodes(D, pos)
nx.draw_networkx_edges(D, pos)
nx.draw_networkx_labels(D, pos)
plt.show()
```
